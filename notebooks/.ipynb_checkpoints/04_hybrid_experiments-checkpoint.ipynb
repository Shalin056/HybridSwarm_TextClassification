```python
# Import Libraries and Setup Paths
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score
import tensorflow as tf
import pickle
from transformers import AutoTokenizer  # Added for BERT tokenization

module_path = r"C:\Users\shali\Documents\shalin\ASU_2nd_SEM\APM 523 Optimization\APM523_HybridSwarm_TextClassification\src"
if module_path not in sys.path:
    sys.path.append(module_path)

print("SRC Path Exists:", os.path.exists(module_path))
print("SRC Path:", module_path)

try:
    from models import build_lstm_model, BertClassifier
    print("Successfully imported models!")
except ImportError as e:
    print("Error importing models:", e)

np.random.seed(42)
tf.random.set_seed(42)
sns.set(style="whitegrid")
sns.set_palette("husl")
```

```python
# Load Baseline and Swarm Results
output_dir = '../outputs/'
results_dir = os.path.join(output_dir, 'results')
baseline_results_path = os.path.join(results_dir, 'baseline_results.csv')
swarm_results_path = os.path.join(results_dir, 'swarm_optimization_results.csv')

baseline_df = pd.read_csv(baseline_results_path)
swarm_df = pd.read_csv(swarm_results_path)

combined_df = pd.concat([baseline_df, swarm_df[['Model', 'Test_Accuracy', 'Test_F1_Score']].rename(columns={'Test_Accuracy': 'Accuracy', 'Test_F1_Score': 'F1-Score'})], ignore_index=True)

print("Baseline Results:")
print(baseline_df)
print("\nSwarm Optimization Results:")
print(swarm_df)
print("\nCombined Results:")
print(combined_df)
```

```python
# Visualize Accuracy Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=combined_df)
plt.title('Test Accuracy Comparison: Baseline vs. Swarm-Optimized Models', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)
for i, v in enumerate(combined_df['Accuracy']):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom')
plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'accuracy_comparison.png'))
plt.show()
```

```python
# Visualize F1-Score Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='F1-Score', data=combined_df)
plt.title('Test F1-Score Comparison: Baseline vs. Swarm-Optimized Models', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('F1-Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)
for i, v in enumerate(combined_df['F1-Score']):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom')
plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'f1_score_comparison.png'))
plt.show()
```

```python
models_dir = os.path.join(output_dir, 'models')
baseline_lstm_path = os.path.join(models_dir, 'baseline_lstm_default.keras')
baseline_cnn_path = os.path.join(models_dir, 'baseline_cnn_default.keras')
baseline_bert_path = os.path.join(models_dir, 'baseline_bert_default.keras')
pso_lstm_path = os.path.join(models_dir, 'pso_lstm.keras')
gwo_lstm_path = os.path.join(models_dir, 'gwo_lstm.keras')
hybrid_lstm_path = os.path.join(models_dir, 'hybrid_lstm.keras')

processed_dir = '../data/processed/'
with open(os.path.join(processed_dir, 'test_tfidf.pkl'), 'rb') as f:
    X_test_tfidf = pickle.load(f)
test_df = pd.read_csv(os.path.join(processed_dir, 'test_preprocessed.csv'))

# Use a subset of test data for faster validation
subset_size = 1000  # Reduced from 7600
test_subset_df = test_df.sample(n=subset_size, random_state=42)
X_test_tfidf_subset = X_test_tfidf[test_subset_df.index]
y_test_subset = test_subset_df['Class Index'].values - 1
y_test_subset_cat = tf.keras.utils.to_categorical(y_test_subset, num_classes=4)

# Tokenize test subset for BERT
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
max_length = 64
test_encodings = tokenizer(test_subset_df['processed_text'].tolist(), max_length=max_length, padding='max_length', 
                           truncation=True, return_tensors='tf')
X_test_bert_subset = {'input_ids': test_encodings['input_ids'], 'attention_mask': test_encodings['attention_mask']}

models = {
    'LSTM': tf.keras.models.load_model(baseline_lstm_path),
    'CNN': tf.keras.models.load_model(baseline_cnn_path),
    'BERT (Subset)': tf.keras.models.load_model(baseline_bert_path, custom_objects={'BertClassifier': BertClassifier}),
    'PSO-LSTM': tf.keras.models.load_model(pso_lstm_path),
    'GWO-LSTM': tf.keras.models.load_model(gwo_lstm_path),
    'Hybrid-PSO-GWO-LSTM': tf.keras.models.load_model(hybrid_lstm_path)
}

results_validation = []
batch_size = 32  # Explicit batch size for BERT prediction
for name, model in models.items():
    if name == 'BERT (Subset)':
        # Predict in batches for BERT
        dataset = tf.data.Dataset.from_tensor_slices(X_test_bert_subset).batch(batch_size)
        y_pred = model.predict(dataset, verbose=0)
    else:
        y_pred = model.predict(X_test_tfidf_subset, verbose=0)
    y_pred_labels = np.argmax(y_pred, axis=1)
    acc = accuracy_score(y_test_subset, y_pred_labels)
    f1 = f1_score(y_test_subset, y_pred_labels, average='weighted')
    results_validation.append({'Model': name, 'Accuracy': acc, 'F1-Score': f1})

validation_df = pd.DataFrame(results_validation)
print("Validation of Loaded Models on Test Subset (1000 samples):")
print(validation_df)

combined_df_validated = combined_df.merge(validation_df, on='Model', suffixes=('_Saved', '_Validated'), how='left')
print("\nConsistency Check (Saved vs. Validated on Subset):")
print(combined_df_validated)
```

```python
# Cell 6: Analyze and Save Conclusions
print("Summary Statistics:")
print(combined_df.describe())

best_model = combined_df.loc[combined_df['Accuracy'].idxmax()]
print("\nBest Performing Model:")
print(best_model)

combined_df.to_csv(os.path.join(results_dir, 'combined_results.csv'), index=False)
print(f"Combined results saved to {results_dir}/combined_results.csv")
```

```python
import os
import sys
import numpy as np
import tensorflow as tf
import pickle
from transformers import AutoTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Add src to path
module_path = r"C:\Users\shali\Documents\shalin\ASU_2nd_SEM\APM 523 Optimization\APM523_HybridSwarm_TextClassification\src"
if module_path not in sys.path:
    sys.path.append(module_path)

from models import BertClassifier  # Ensure models.py has the updated BertClassifier

# Paths
processed_dir = '../data/processed/'
models_dir = '../outputs/models/'
tfidf_vectorizer_path = os.path.join(processed_dir, 'tfidf_vectorizer.pkl')
gwo_lstm_path = os.path.join(models_dir, 'gwo_lstm.keras')
bert_path = os.path.join(models_dir, 'baseline_bert_default.keras')

# Load TF-IDF vectorizer
with open(tfidf_vectorizer_path, 'rb') as f:
    vectorizer = pickle.load(f)

# Load models
gwo_lstm_model = tf.keras.models.load_model(gwo_lstm_path)
bert_model = tf.keras.models.load_model(bert_path, custom_objects={'BertClassifier': BertClassifier})

# Load BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
max_length = 64

# Class labels (adjust based on your dataset)
class_labels = {0: "World", 1: "Sports", 2: "Business", 3: "Sci/Tech"}

def preprocess_text(text, for_bert=False):
    if for_bert:
        encodings = tokenizer([text], max_length=max_length, padding='max_length', truncation=True, return_tensors='tf')
        return {'input_ids': encodings['input_ids'], 'attention_mask': encodings['attention_mask']}
    else:
        tfidf = vectorizer.transform([text]).toarray()
        return tfidf

def predict_text(model, text, is_bert=False):
    preprocessed = preprocess_text(text, for_bert=is_bert)
    pred = model.predict(preprocessed, verbose=0)
    class_idx = np.argmax(pred, axis=1)[0]
    return class_labels[class_idx], pred[0][class_idx]

# Interactive loop
print("Text Classification Demo (Enter 'quit' to exit)")
while True:
    user_input = input("Enter a news headline or text: ").strip()
    if user_input.lower() == 'quit':
        print("Exiting...")
        break
    if not user_input:
        print("Please enter some text!")
        continue

    # Predict with GWO-LSTM
    gwo_pred, gwo_prob = predict_text(gwo_lstm_model, user_input, is_bert=False)
    print(f"GWO-LSTM Prediction: {gwo_pred} (Confidence: {gwo_prob:.4f})")

    # Predict with BERT
    bert_pred, bert_prob = predict_text(bert_model, user_input, is_bert=True)
    print(f"BERT Prediction: {bert_pred} (Confidence: {bert_prob:.4f})")
    print("-" * 50)
```
